## Franchising Model Simulation
------------------------------------------------

```python
# Parameters

SalesGrowth = 0.25
GPGrowth = 0.12
InitialGP = 0.2
Charge = .25
annual = 365
equipment = 400000
amortization = 6
HQ = 12000000
stores = [50,100,150,200]
```

```python
franchise = pd.DataFrame()
```

```python
franchise['PreDaySales'] = np.arange(6000,13500,500) 
```

```python
franchise['PostDaySales'] = franchise['PreDaySales'] * (1 + SalesGrowth)
```

```python
franchise['InitialGP'] = franchise['PreDaySales'] * InitialGP
```

```python
franchise['NewGP'] = franchise['PostDaySales'] * (InitialGP + GPGrowth)
```

```python
franchise['NewProfitOwner'] =  franchise['NewGP'] * (1 - Charge)
```

```python
franchise['7ElevenProfit'] =  franchise['NewGP'] * Charge
```

```python
franchise['OwnerIncrease'] = (franchise['NewProfitOwner']/franchise['InitialGP']) - 1
```

```python
franchise['NewProfitOwnerAnnual'] = franchise['NewProfitOwner'] * annual
```

```python
franchise['7ElevenProfitAnnual'] = franchise['7ElevenProfit'] * annual
```

```python
franchise['7ElevenCost@50'] = franchise['7ElevenProfitAnnual']- ((HQ / stores[0]) + (equipment/ amortization))
franchise['7ElevenCost@100'] = franchise['7ElevenProfitAnnual']- ((HQ / stores[1]) + (equipment/ amortization))
franchise['7ElevenCost@150'] = franchise['7ElevenProfitAnnual']- ((HQ / stores[2]) + (equipment/ amortization))
franchise['7ElevenCost@200'] = franchise['7ElevenProfitAnnual']- ((HQ / stores[3]) + (equipment/ amortization))
```

```python
import itertools
salesup = np.arange(200,305,5)
gpup = np.arange(250,355,5)
precios = np.arange(8000,14300,300)
ownergp = 0.25
charges = [200,210,220,230,240,250,260,270,280,290,300]
royalty = 0.015

simuls = {}

for charge in charges:
    simulaowner = []
    simula7eleven = []
    simulaoldowner = []
    ventas = []
    gps = []
    precioss = []
    for sales, gp,precio in itertools.product(salesup, gpup,precios):
        owner = ((((precio * (1 +(sales/1000)))*(1 - royalty))* (gp/1000))*(1- (charge/1000)))* annual
        ownerold = (precio * ownergp)* annual
        seveneleven = (((precio * (1 +(sales/1000)))* (gp/1000))*(charge/1000)) * annual
        simulaowner.append(owner)
        simulaoldowner.append(ownerold)
        simula7eleven.append(seveneleven)
        ventas.append(sales)
        gps.append(gp)
        precioss.append(precio)
    simulation = pd.DataFrame()
    simulation['Owner New $'] = simulaowner
    simulation['7Eleven $'] = simula7eleven
    simulation['Owner old'] = simulaoldowner
    simulation['Sales Delta'] = ventas
    simulation['GPs'] = gps
    simulation['Sales'] = precio
    simulation['Delta $'] = simulation['Owner New $'] - simulation['Owner old']
    simulation['Delta %'] = (simulation['Owner New $'] / simulation['Owner old'])
    simuls[charge] = simulation
```

```python
simula = {}
stats = {}

for charge in charges:
    simula[charge] = simuls[charge]
for charge in charges:
    simula[charge] = simula[charge].drop(['Sales Delta','GPs','Sales'],axis=1)
for charge in charges:
    stats[charge] = simula[charge].describe()
    stats[charge] = stats[charge].applymap('${:,.2f}'.format)
```

```python
import numpy as np
import pandas as pd
import itertools


class Store:
    def __init__(self,MinSales =8000,
                 MaxSales=14000,
                 GrossProfit=0.25):
        self.MinSales = MinSales
        self.MaxSales = MaxSales
        self.GrossProfit = GrossProfit
        
        
class Eleven(Store):
    annual = 365
    def __init__(self,MinSales = 8000,
                 MaxSales = 14000,
                 GrossProfit = 0.25,
                 Royalty = 0.015,
                 Min_Increase = 0.20, 
                 Max_Increase = 0.30,
                 GPup = 0.10,
                 GPdown = 0.0,
                 Intervals = 20,
                 multiplier = 1000,
                 royalty = 0.015):
        super().__init__(MinSales,MaxSales,GrossProfit)
        self.Royalty = Royalty
        self.MinIncrease = Min_Increase
        self.MaxIncrease = Max_Increase
        self.GPup = GPup
        self.GPdown = GPdown
        self.Intervals = Intervals
        self.multiplier = multiplier
        self.royalty = royalty
        
    def step(self,Maximo,Minimo,Paso):
        step = (Maximo - Minimo)/Paso
        return step
    
    def simulacion(self,charge = 0.25):
        salesstep = self.step(self.MaxIncrease,self.MinIncrease,self.Intervals)
        
        salesup = np.arange(self.MinIncrease * self.multiplier,
                            (self.MaxIncrease + salesstep)* self.multiplier,
                            salesstep* self.multiplier)
        gpstep = self.step(self.GPup,self.GPdown,self.Intervals)
        
        gpup = np.arange((self.GrossProfit + self.GPdown) * self.multiplier,
                         ((self.GrossProfit + self.GPup)+gpstep)* self.multiplier,
                          gpstep*self.multiplier)
        
        preciosstep = self.step(self.MaxSales,self.MinSales,self.Intervals)
        
        precios = np.arange(self.MinSales,
                            (self.MaxSales + preciosstep),
                            preciosstep)
               
        simulaowner = []
        simula7eleven = []
        simulaoldowner = []
        ventas = []
        gps = []
        precioss = []
        for sales, gp,precio in itertools.product(salesup, gpup,precios):
            owner = ((((precio * (1 +(sales/1000)))*(1 - self.royalty))* (gp/1000))*(1- (charge)))* self.annual
            ownerold = (precio * self.GrossProfit)* self.annual
            seveneleven = (((precio * (1 +(sales/1000)))* (gp/1000))*(charge)) * self.annual
            simulaowner.append(owner)
            simulaoldowner.append(ownerold)
            simula7eleven.append(seveneleven)
            ventas.append(sales)
            gps.append(gp)
            precioss.append(precio)
        self.simulation = pd.DataFrame()
        self.simulation['Owner New $'] = simulaowner
        self.simulation['7Eleven $'] = simula7eleven
        self.simulation['Owner old'] = simulaoldowner
        self.simulation['Sales Delta'] = ventas
        self.simulation['GPs'] = gps
        self.simulation['Sales'] = precio
        self.simulation['Delta $'] = self.simulation['Owner New $'] - self.simulation['Owner old']
        self.simulation['Delta %'] = (self.simulation['Owner New $'] / self.simulation['Owner old'])
        
        return self.simulation.describe()
    
    def valores(self):
        print(self.MaxIncrease,self.MinIncrease,self.Intervals)
        print(self.multiplier,self.GPup,self.GPdown,self.Intervals)
        print(self.GrossProfit,self.MaxSales,self.MinSales)
        print(self.royalty,self.annual, self.GrossProfit)
```
