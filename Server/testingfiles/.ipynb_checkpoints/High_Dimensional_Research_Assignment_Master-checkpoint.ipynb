# High Dimensional Data

All of our work to this point is building dimensions on the bases of linear algebra and function analysis. This section involves studying some of the specific characteristics of high dimensionality that affect your analysis and data preparation.

1. High dimensional data cannot be intuited without modification
2. As dimensions increase, the number of choices explode
3. Selecting good data gets more difficult
4. In extremely high dimensional cases, only automated methods can be used.

# Assignment in 8D

* Calculate the centroid of the survivors.
* Calculate the centroid of the casualties.
* Calculate the average distance between each survivor.
* Calculate the average distance between each casualty.
* Calculate the distance between the two centroids. Which is farthest? Which is nearest?


```python
pd.options.mode.chained_assignment = None

cen_sur = titanic[titanic['survived']==1]
cen_sur.drop(['survived','alone'],axis =1,inplace = True)
cen_sur.head()
```

```python
cen_cas = titanic[titanic['survived']==0]
cen_cas.drop(['survived','alone'],axis =1,inplace = True)
cen_cas.head()
```

```python
def centeroidnp(frame):
    vector = []
    largo = frame.shape[0]
    for col in frame.columns:
        sum_d = frame[col].sum()
        cen_d = sum_d/largo
        vector.append(cen_d)
        vector1 = np.asarray(vector)
    return vector1
```

```python
grupos = [['Survivors:',cen_sur],['Casualties:',cen_cas]]
centroids =[]
for i in grupos:
    centro = centeroidnp(i[1])
    centroids.append(centro)
    print('{}\n{}'.format(i[0], centro))
```

```python
import itertools as it
```

```python
cen_sur.reset_index(inplace=True, drop = True)
cen_cas.reset_index(inplace=True, drop = True)
```

```python
cen_sur_comb = cen_sur.index.tolist()
cen_cas_comb = cen_cas.index.tolist()
```

```python
cen_sur_res = list(it.combinations(cen_sur_comb,2))
cen_cas_res = list(it.combinations(cen_cas_comb,2))

print(len(cen_sur_res))
print(len(cen_cas_res))
```

```python
cen_sur.head()
```

```python
cen_cas.head()
```

```python
cen_sur.iloc[0].values
```

```python
def  euclidean_distance(x1,x2):
    dist = 0
    for i in range(len(x1)):
        dist+=(x1[i] - x2[i])**2 
    return dist**0.5
```

```python
acu_cen_sur = 0
for res in cen_sur_res:
    distancia = euclidean_distance(cen_sur.iloc[res[0]].values,cen_sur.iloc[res[1]].values)
    acu_cen_sur += distancia
    
pro_cen_sur = acu_cen_sur/len(cen_sur_res)
```

```python
acu_cas_sur = 0
for res in cen_cas_res:
    distancia = euclidean_distance(cen_cas.iloc[res[0]].values,cen_cas.iloc[res[1]].values)
    acu_cas_sur += distancia
    
pro_cen_cas = acu_cas_sur/len(cen_cas_res)
```

```python
print(pro_cen_sur)
print(pro_cen_cas)
```
